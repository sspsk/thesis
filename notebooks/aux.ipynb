```python
import torch
import matplotlib.pyplot as plt

import torch.nn.functional as F
```

```python
x = torch.linspace(-5,5,100)
```

```python
relu = F.relu(x)
sigm = F.sigmoid(x)
tanh = F.tanh(x)
lrelu = F.leaky_relu(x,negative_slope=0.1)
```

```python
fig,axs = plt.subplots(1,4,figsize=(20,5))

lw=2.0

axs[0].plot(x,sigm,linewidth=lw)
axs[0].set_ylim(0,1)
axs[0].set_title('Sigmoid')
axs[1].plot(x,tanh,linewidth=lw)
axs[1].set_ylim(-1,1)
axs[1].set_title('Tanh')
axs[2].plot(x,relu,linewidth=lw)
axs[2].set_ylim(-1,5)
axs[2].set_title('ReLU')
axs[3].plot(x,lrelu,linewidth=lw)
axs[3].set_ylim(-1,5)
axs[3].set_title('Leaky ReLU')

for i in range(4):
    axs[i].grid()


plt.savefig('activations.jpg')
```

```python
x = torch.linspace(-2,2,100)
y = x**2
y_tilde = x**2 + 0.5*torch.rand_like(x)

class1 = torch.randn(2,100) + torch.tensor([[2],[2]])
class2 = torch.randn(2,100) + torch.tensor([[0],[0]])
xs = torch.linspace(-4,4,100)
ys = -xs + 1.5

fig,axs = plt.subplots(1,2,figsize=(10,5))

axs[0].grid()
axs[0].scatter(class1[0],class1[1],c='red',label='class 1')
axs[0].scatter(class2[0],class2[1],c='green',label='class 2')
axs[0].plot(xs,ys,color='black',label='Decision Boundary',linewidth=2.0,linestyle='dashed')
axs[0].legend()
axs[0].set_axisbelow(True)
axs[0].set_title('Ταξινόμηση')

axs[1].grid()
axs[1].scatter(x,y_tilde,label='Data')
axs[1].plot(x,y,color='black',label='Predicted values',linewidth=2.0,linestyle='dashed')
axs[1].legend()
axs[1].set_axisbelow(True)
axs[1].set_title('Παλινδρόμηση')
```
